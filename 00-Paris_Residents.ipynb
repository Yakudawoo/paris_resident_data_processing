## Task 10: Generate district summary report

Create a function called `generate_district_report` that takes a list of residents and a district number. The function should:

1. Filter residents by the specified district
2. Calculate the average age in that district
3. Count total residents
4. Return a formatted string

The average should be rounded to 1 decimal place.

Test with district 11 and district 2.

````python
# Write your function here
def generate_district_report(residents, district_number):
    # list comprehension district_residents pour filtrer les résidents par district
    district_residents = [resident for resident in residents if resident["district"] == district_number]
# le premier resident (variable) me permet de faire ceci  :for resident in district_residents
# dans avg_age
    if not district_residents:
        return f"No residents found in district {district_number}."

    avg_age = round(sum(resident["age"] for resident in district_residents) / len(district_residents), 1)
    count = len(district_residents)

    return f"District {district_number}: {count} residents, average age {avg_age}"


generate_district_report(residents, 11)  # Should return "District 11: 1 residents, average age 29.0"
print(generate_district_report(residents, 11))
generate_district_report(residents, 2)   # Should return "District 2: 1 residents, average age 28.0"
print(generate_district_report(residents, 2))
generate_district_report(residents, 5)  # Should return "No residents found in district 5"
print(generate_district_report(residents, 5))
````

## Task 11: Validate zipcode format

Paris zipcodes must follow the format 75XXX where X is any digit. Create a function called `validate_paris_zipcode` that takes a zipcode string and returns `True` if it matches the Paris format, `False` otherwise.

Requirements:
- Must start with "75"
- Must be exactly 5 characters long
- All 5 characters must be digits

````python
# Write your function here
def validate_paris_zipcode(zipcode):
    if zipcode.startswith("75") and len(zipcode) == 5 and zipcode.isdigit():
        if 75001 <= int(zipcode) <= 75020:
            return True
    return False    
# Test cases
print(f"75011 valid: {validate_paris_zipcode('75011')}")  # Should return True
print(f"75999 valid: {validate_paris_zipcode('75999')}")  # Should return False
print(f"92100 valid: {validate_paris_zipcode('92100')}")    # Should return False
print(f"7501 valid: {validate_paris_zipcode('7501')}")  # Should return False
print(f"75abc valid: {validate_paris_zipcode('75abc')}")  # Should return False
````

## Task 14: Calculate age distribution

Create a function called `calculate_age_distribution` that takes a list of residents and returns a dictionary showing how many residents fall into each age category:

- "young": ages 0-30
- "middle": ages 31-50
- "senior": ages 51+

Test with the `residents` list from the setup cell.

````python
# Write your function here
def calculate_age_distribution(residents):
    age_categories = {
        "young": 0,
        "middle": 0,
        "senior": 0
    }
    for resident in residents:
        age = resident["age"]
        if age <= 30:
            age_categories["young"] += 1
        elif 31 <= age <= 50:
            age_categories["middle"] += 1
        else:
            age_categories["senior"] += 1
    return age_categories

age_distribution = calculate_age_distribution(residents)
print(f"Age Distribution: ")  # Should return a dictionary with counts per
print(f" Young (<=30): {age_distribution['young']}")
print(f" Middle (31-50): {age_distribution['middle']}")
print(f" Senior (>50): {age_distribution['senior']}") 
````

## Task 15: Find oldest and youngest residents

Create a function called `find_age_extremes` that takes a list of residents and returns a dictionary with two keys:

````python
# Write your function here
def find_age_extremes(residents):
    if not residents:
        return {"youngest": None, "oldest": None}
    
    ages = [resident["age"] for resident in residents]
    # list comprehension pour extraire les âges des résidents dans une nle liste ages
    # équivalent verbeux 
    # ages = []
    # for resident in residents:
    #     ages.append(resident["age"])
    # resident["age"] accède à la valeur associée à la clé "age" dans chaque dictionnaire resident
    # puis retourne des valeurs d'âge individuelles qui sont collectées dans la liste ages.
    min_age = min(ages)
    max_age = max(ages)
    
    return {"youngest": min_age, "oldest": max_age, "ages": ages}


find_age_extremes(residents)  # Should return (28, 52)
print(f"Ages list: {find_age_extremes(residents)['ages']}")  # Should return [34, 28, 45, 52, 29]
print(f"Youngest resident: {find_age_extremes(residents)['youngest']}")  # Should return (28)
print(f"Oldest resident: {find_age_extremes(residents)['oldest']}")  # Should return (52)


# example with empty list
empty_list = []
print(f"Age extremes in empty list: {find_age_extremes(empty_list)}")  # Should return (None, None)
````
